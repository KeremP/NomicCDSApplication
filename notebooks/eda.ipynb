Reaction counts / most engaged messages

```python
all_data.drop(
    all_data.loc[all_data['channel_name'] == "roles"].index,
    inplace=True
)
```

```python
def get_reaction_count(x: str):
    if type(x) is not str: return 0
    x = x.replace("'",'"')
    x_split = x.split("emoji")
    counts = 0
    for s in x_split:
        if "count" not in s: continue
        try:
            counts+=int(
                    s.split('count":')[1].split(",")[0].strip()
                )
        except:
            continue
    return counts
```

```python
all_data['reaction_count'] = all_data['reactions'].apply(lambda x: get_reaction_count(x))
```

```python
sum(all_data['reaction_count'])
```

```python
reactions = all_data.sort_values(by=['reaction_count'], ascending=False)
```

```python
unique_users = joins['author']
```

```python
def get_id_username(x: str):
    splits = x.split("',")
    id = splits[0].split(": ")[1].replace("'","")
    username = splits[1].split(": ")[1].replace("'","")
    return id, username
```

```python
def get_id(x: str):
    id, _ = get_id_username(x)
    return id

def get_username(x: str):
    _, username = get_id_username(x)
    return username
```

```python
ids = joins['author'].apply(get_id)
```

```python
usernames = joins['author'].apply(get_username)
```

```python
users = pd.DataFrame({"user_id":ids,"username":usernames, "join_time": joins['timestamp']})
```

```python
len(users['user_id'].unique())
```

```python
users = users.sort_values(by="join_time")
```

```python
users = users.drop_duplicates(subset=['user_id'], keep="first")
```

```python
all_data['user_id'] = all_data['author'].apply(get_id)
all_data['user_name'] = all_data['author'].apply(get_username)
```

```python
merged_data = all_data.merge(users[['user_id','join_time']], on="user_id", how="inner")
```

```python
merged_data_ex_joins = merged_data.loc[merged_data['channel_name']!="joins"]
```

```python
merged_data_ex_joins['time_to_comm'] = (merged_data_ex_joins['timestamp'] - merged_data_ex_joins['join_time']) / np.timedelta64(1, 'm')
```

```python
sorted_msgs = merged_data_ex_joins.sort_values(by="timestamp")
```

```python
sorted_msgs.drop_duplicates(subset=['user_id'], keep="first", inplace=True)
```

```python
sorted_msgs = sorted_msgs.loc[sorted_msgs['time_to_comm'] > 0]
```

```python
sorted_msgs['time_to_comm'].median()
```
